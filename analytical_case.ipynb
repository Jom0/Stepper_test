```python
def density(x,lbd_t):
    return 1/(2*np.pi*np.linalg.det(lbd_t[1]))*np.exp(-0.5*(np.array([x-lbd_t[0]])@np.linalg.inv(lbd_t[1])@np.array([x-lbd_t[0]]).T)[0,0])
```

```python
def base_density(x,lbd_base):
    return lbd_t[0]*lbd_t[1]*np.exp(-lbd_t[0]*x[0]-lbd_t[1]*x[1])
```

```python
def limit_state(x):
    return 100-1.1*x
```

```python
from itertools import chain, combinations

def powerset(iterable):
    "powerset([1,2,3]) --> (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return list(chain.from_iterable(combinations(s, r) for r in range(1,len(s)+1)))

powerset([0,1,2])
```
